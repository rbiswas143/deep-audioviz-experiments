```python
# Read the tracks
import librosa
import os

num_tracks = 100

def read_tracks():
    data=[]
    srs=[]
    track_count=0
    for root, dirs, files in os.walk('../fma/fma_small'):
        for file_ in files:
            if 'mp3' in file_:
                if track_count >= num_tracks:
                    return data, srs
                file_path = os.path.join(root, file_)
                print 'Rading file %d of %d: %s' % (track_count+1, num_tracks, file_path)
                d,sr = librosa.load(file_path)
                data.append(d)
                srs.append(sr)
                track_count+=1
    return data, srs

data, srs = read_tracks()
```
